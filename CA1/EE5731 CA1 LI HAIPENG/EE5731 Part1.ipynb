```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2


srcImg = cv2.imread('Img/einstein.png')
srcImg = cv2.cvtColor(srcImg,cv2.COLOR_BGR2GRAY)

#def different kernel
sobel_kernel_vert = np.array([[-1,0,1],
                              [-2,0,2],
                              [-1,0,1]])

sobel_kernel_hori = np.array([[-1,-2,-1],
                              [0,0,0],
                              [1,2,1]])

guassian_kernel_3 = 1/16 * np.array([[1,2,1],
                                     [2,4,2],
                                     [1,2,1]])

guassian_kernel_5 = 1/256 * np.array([[1,4,6,4,1],
                                     [4,16,24,16,4],
                                     [6,24,36,24,6],
                                     [4,16,24,16,4],
                                     [1,4,6,4,1]])

def generate_outImg(srcImg,kernel):
    m = srcImg.shape[0]
    n = srcImg.shape[1]
    outImg = np.zeros([m-kernel.shape[0]+1, n-kernel.shape[1]+1])
    return outImg

def conv_2d(src,kernel):
    out = generate_outImg(src,kernel)
    conv(src,out,kernel)
    return out

def conv(src,out,kernel):
    for i in range(out.shape[0]):
        for j in range(out.shape[1]):
            src_pitch = src[i:i+kernel.shape[0], j:j+kernel.shape[1]]
            pixel_count = kernel.size
            pixel_sum = 0
            srcflat = src_pitch.flatten()
            kernelflat = kernel.flatten()
            for k in range(pixel_count):
                pixel_sum += srcflat[k]*kernelflat[k]
                
            out[i, j] = pixel_sum

#====================================================
plt.figure(figsize=(20,20))
plt.subplot(511)
plt.title('einstein')
plt.imshow(srcImg,cmap="gray",vmin = 0,vmax = 255)
    
sobel_vImg = conv_2d(srcImg,sobel_kernel_vert)
plt.subplot(512)
plt.title('sobel_kernel_vert')
plt.imshow(sobel_vImg,cmap="gray",vmin = 0,vmax = 255)

sobel_hImg = conv_2d(srcImg,sobel_kernel_hori)
plt.subplot(513)
plt.title('sobel_kernel_hori')
plt.imshow(sobel_hImg,cmap="gray",vmin = 0,vmax = 255)

guass3Img = conv_2d(srcImg,guassian_kernel_3)
plt.subplot(514)
plt.title('guassian_kernel_3')
plt.imshow(guass3Img,cmap="gray",vmin = 0,vmax = 255)


guass5Img = conv_2d(srcImg,guassian_kernel_5)
plt.subplot(515)
plt.title('guassian_kernel_5')
plt.imshow(guass5Img,cmap="gray",vmin = 0,vmax = 255)
plt.show()
```

```python
#generate the integral image
def integral(src):
    inteImg = np.zeros([src.shape[0]+1, src.shape[1]+1])#first row and column are 0
    for i in range(1, src.shape[0]):
        for j in range(1, src.shape[1]):
            inteImg[i][j] = src[i-1][j-1] + inteImg[i-1][j] + inteImg[i][j-1] - inteImg[i-1][j-1]
            
    return inteImg

#basic x2 mask is [-1,1]
def haar_x2(src,inteImg,scale):
    w = 2 * scale
    h = 1 * scale
    haarImg = np.zeros([src.shape[0]-h+1, src.shape[1]-w+1])
    for i in range(0, haarImg.shape[0]-1):
        for j in range(0, haarImg.shape[1]-1):
            m = h + i
            n = w + j
            haar_all = inteImg[m][n]+inteImg[i][j]-inteImg[m][j]-inteImg[i][n]
            haar_black = inteImg[m][n-int(w/2)]+inteImg[i][j]-inteImg[m][j]-inteImg[i][n-int(w/2)]
            haarImg[i][j] = haar_all - haar_black*2
    return haarImg

#basic y2 mask is [[-1],[1]]
def haar_y2(src,inteImg,scale):
    w = 1 * scale
    h = 2 * scale
    haarImg = np.zeros([src.shape[0]-h+1, src.shape[1]-w+1])
    for i in range(0, haarImg.shape[0]-1):
        for j in range(0, haarImg.shape[1]-1):
            m = h + i
            n = w + j
            haar_all = inteImg[m][n]+inteImg[i][j]-inteImg[m][j]-inteImg[i][n]
            haar_black = inteImg[m-int(h/2)][n]+inteImg[i][j]-inteImg[m-int(h/2)][j]-inteImg[i][n]
            haarImg[i][j] = haar_all - haar_black*2
    return haarImg

#basic x3 mask is [1,-1,1]
def haar_x3(src,inteImg,scale):
    w = 3 * scale
    h = 1 * scale
    haarImg = np.zeros([src.shape[0]-h+1, src.shape[1]-w+1])
    for i in range(0, haarImg.shape[0]-1):
        for j in range(0, haarImg.shape[1]-1):
            m = h + i
            n = w + j
            haar_all = inteImg[m][n]+inteImg[i][j]-inteImg[m][j]-inteImg[i][n]
            haar_black = inteImg[m][n-int(w/3)]+inteImg[i][j+int(w/3)]-inteImg[m][j+int(w/3)]-inteImg[i][n-int(w/3)]
            haarImg[i][j] = haar_all - haar_black*3
    return haarImg

#basic y3 mask is [[1],[-1],[1]]
def haar_y3(src,inteImg,scale):
    w = 1 * scale
    h = 3 * scale
    haarImg = np.zeros([src.shape[0]-h+1, src.shape[1]-w+1])
    for i in range(0, haarImg.shape[0]-1):
        for j in range(0, haarImg.shape[1]-1):
            m = h + i
            n = w + j
            haar_all = inteImg[m][n]+inteImg[i][j]-inteImg[m][j]-inteImg[i][n]
            haar_black = inteImg[m-int(h/3)][n]+inteImg[i+int(h/3)][j]-inteImg[m-int(h/3)][j]-inteImg[i+int(h/3)][n]
            haarImg[i][j] = haar_all - haar_black*3
    return haarImg

#basic x2y2 mask is [[-1,1],[1,-1]]
def haar_x2y2(src,inteImg,scale):
    w = 2 * scale
    h = 2 * scale
    haarImg = np.zeros([src.shape[0]-h+1, src.shape[1]-w+1])
    for i in range(0, haarImg.shape[0]-1):
        for j in range(0, haarImg.shape[1]-1):
            m = h + i
            n = w + j
            haar_all = inteImg[m][n]+inteImg[i][j]-inteImg[m][j]-inteImg[i][n]
            haar_black = inteImg[m][n]+2*inteImg[m-int(h/2)][n-int(w/2)]+inteImg[i][j]-inteImg[m][n-int(w/2)]-inteImg[m-int(h/2)][n]-inteImg[m-int(h/2)][j]-inteImg[i][n-int(w/2)]
            haarImg[i][j] = haar_all - haar_black*2           
    return haarImg
#==========================================
inteImg = integral(srcImg)
scale = 1
haarx2Img = haar_x2(srcImg,inteImg,scale)
haary2Img = haar_y2(srcImg,inteImg,scale)
haarx3Img = haar_x3(srcImg,inteImg,scale)
haary3Img = haar_y3(srcImg,inteImg,scale)
haarx2y2Img = haar_x2y2(srcImg,inteImg,scale)

plt.figure(figsize=(20,20))
plt.subplot(321)
plt.title('haarx2,scale = 1')
plt.imshow(haarx2Img,cmap="gray",vmin = 0,vmax = 255)
plt.subplot(322)
plt.title('haary2,scale = 1')
plt.imshow(haary2Img,cmap="gray",vmin = 0,vmax = 255)
plt.subplot(323)
plt.title('haarx3,scale = 1')
plt.imshow(haarx3Img,cmap="gray",vmin = 0,vmax = 255)
plt.subplot(324)
plt.title('haary3,scale = 1')
plt.imshow(haary3Img,cmap="gray",vmin = 0,vmax = 255)
plt.subplot(325)
plt.title('haarx2y2,scale = 1')
plt.imshow(haarx2y2Img,cmap="gray",vmin = 0,vmax = 255)
```

## 3.Write your conclusions on the outputs of each of the kernels, and explain how the size of the kernel influences the outputs.

#### Sobel kernel can detect horizontal and vertical line in the image, which usually used at edge detection.
#### Guassian kernel will blur the image, and it is also used to reduce the image noise.
#### Haar-like x2 and y2 masks are used to find the edge features, and indicate where the boundary between the dark and light areas. Haar-like x3 and y3 masks are used to find the line features. Haar-like x2y2 is used to find the four-rectangle features.
#### The different size of the kernel can find more features with different scale. The small kernel can find some details about the image, so the features in the outputs are quite small. While the big kernel can show some bigger features in the outputs. Besides, the size of kernel will also influence the size of the output images.
