## Extract SIFT keypoints from im01.jpg using any publicly available SIFT code. Show the keypoints on the images.

```python
#delete all variables
%reset -f

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2

img1 = cv2.imread('Img/im01.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)

img2 = cv2.imread('Img/im02.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

img3 = cv2.imread('Img/im03.jpg')
img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)

img4 = cv2.imread('Img/im04.jpg')
img4 = cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)

img5 = cv2.imread('Img/im05.jpg')
img5 = cv2.cvtColor(img5,cv2.COLOR_BGR2RGB)


sift = cv2.xfeatures2d.SIFT_create()

#collect kp & des information
(kp1, des1) = sift.detectAndCompute(img1, None)
(kp2, des2) = sift.detectAndCompute(img2, None)
(kp3, des3) = sift.detectAndCompute(img3, None)
(kp4, des4) = sift.detectAndCompute(img4, None)
(kp5, des5) = sift.detectAndCompute(img5, None)

#draw keypoints
f1 = cv2.drawKeypoints(img1, kp1, img1, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
f2 = cv2.drawKeypoints(img2, kp2, img2, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
f3 = cv2.drawKeypoints(img3, kp3, img3, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
f4 = cv2.drawKeypoints(img4, kp4, img4, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
f5 = cv2.drawKeypoints(img5, kp5, img5, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.figure(figsize=(40,40))
plt.subplot(511)
plt.imshow(f1)
plt.subplot(512)
plt.imshow(f2)
plt.subplot(513)
plt.imshow(f3)
plt.subplot(514)
plt.imshow(f4)
plt.subplot(515)
plt.imshow(f5)
plt.show()

```
