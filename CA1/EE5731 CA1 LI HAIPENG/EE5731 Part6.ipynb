```python
#delete all variables
%reset -f

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2
from scipy import linalg

def sift_match(des_left, des_right):
    d = np.zeros(des_right.shape[0])
    flag = np.zeros(des_left.shape[0])
    
    for m in range(0, des_left.shape[0]):
        for n in range(0, des_right.shape[0]):
            d2 = sum((des_left[m]-des_right[n])**2)
            d[n] = np.sqrt(d2)
            
        flag[m] = np.argmin(d)
    return flag   

def getMatch(flag1, flag2):
    match = []
    for i in range(0, len(flag1)):
        for j in range(0, len(flag2)):
            if flag1[i] == j and i == flag2[j]:
                match.append([i,flag1[i]])
    print('len(match)',len(match))
    return match

def drawmatch(img1,img2,match,kp1,kp2):
    splicedImg = np.hstack((img1, img2))        #horizontally spliced img1 + img2
    for i in range(len(match)):
        x1 = kp1[int(match[i][0])].pt[0]
        y1 = kp1[int(match[i][0])].pt[1]
        x2 = kp2[int(match[i][1])].pt[0]
        y2 = kp2[int(match[i][1])].pt[1]
        cv2.line(splicedImg,(int(x1),int(y1)),(int(x2)+img2.shape[1],int(y2)),(255,0,0),1)
    return splicedImg




#=========================================

img1 = cv2.imread('Img/im01.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img1_copy = np.tile(img1, 1)      #copy img 
img2 = cv2.imread('Img/im02.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
img2_copy = np.tile(img2, 1)      #copy img 
img3 = cv2.imread('Img/im03.jpg')
img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
img3_copy = np.tile(img3, 1)      #copy img 
img4 = cv2.imread('Img/im04.jpg')
img4 = cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)
img4_copy = np.tile(img4, 1)      #copy img 
img5 = cv2.imread('Img/im05.jpg')
img5 = cv2.cvtColor(img5,cv2.COLOR_BGR2RGB)
img5_copy = np.tile(img5, 1)      #copy img 
sift = cv2.xfeatures2d.SIFT_create()

#collect kp & des information
(kp1, des1) = sift.detectAndCompute(img1, None)
(kp2, des2) = sift.detectAndCompute(img2, None)
(kp3, des3) = sift.detectAndCompute(img3, None)
(kp4, des4) = sift.detectAndCompute(img4, None)
(kp5, des5) = sift.detectAndCompute(img5, None)
#绘制关键点
f1 = cv2.drawKeypoints(img1_copy, kp1, img1_copy, flags = cv2.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS)
f2 = cv2.drawKeypoints(img2_copy, kp2, img2_copy, flags = cv2.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS)
f3 = cv2.drawKeypoints(img3_copy, kp3, img3_copy, flags = cv2.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS)
f4 = cv2.drawKeypoints(img4_copy, kp4, img4_copy, flags = cv2.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS)
f5 = cv2.drawKeypoints(img5_copy, kp5, img5_copy, flags = cv2.DRAW_MATCHES_FLAGS_NOT_DRAW_SINGLE_POINTS)


flag1t2 = sift_match(des1, des2)
flag2t1 = sift_match(des2, des1)

flag2t3 = sift_match(des2, des3)
flag3t2 = sift_match(des3, des2)

flag3t4 = sift_match(des3, des4)
flag4t3 = sift_match(des4, des3)

flag4t5 = sift_match(des4, des5)
flag5t4 = sift_match(des5, des4)

match12 = getMatch(flag1t2, flag2t1)
match23 = getMatch(flag2t3, flag3t2)
match34 = getMatch(flag3t4, flag4t3)
match45 = getMatch(flag4t5, flag5t4)

sift1 = drawmatch(f1, f2, match12,kp1,kp2)
sift2 = drawmatch(f2, f3, match23,kp2,kp3)
sift3 = drawmatch(f3, f4, match34,kp3,kp4)
#sift4 = drawmatch(f4, f5, match45,kp4,kp5)


plt.figure(figsize=(20,20))
plt.subplot(411)
plt.imshow(sift1)
plt.subplot(412)
plt.imshow(sift2)
plt.subplot(413)
plt.imshow(sift3)
# plt.subplot(414)
# plt.imshow(sift4)
plt.show()
```

```python
#from match choose 4 keypoint pairs
def choosepoint(kp1,kp2,match,n1,n2,n3,n4):
    x1 = kp1[int(match[n1][0])].pt[0]
    y1 = kp1[int(match[n1][0])].pt[1]
    x11 = kp2[int(match[n1][1])].pt[0]
    y11 = kp2[int(match[n1][1])].pt[1]
    
    x2 = kp1[int(match[n2][0])].pt[0]
    y2 = kp1[int(match[n2][0])].pt[1]
    x22 = kp2[int(match[n2][1])].pt[0]
    y22 = kp2[int(match[n2][1])].pt[1]
    
    x3 = kp1[int(match[n3][0])].pt[0]
    y3 = kp1[int(match[n3][0])].pt[1]
    x33 = kp2[int(match[n3][1])].pt[0]
    y33 = kp2[int(match[n3][1])].pt[1]
    
    x4 = kp1[int(match[n4][0])].pt[0]
    y4 = kp1[int(match[n4][0])].pt[1]
    x44 = kp2[int(match[n4][1])].pt[0]
    y44 = kp2[int(match[n4][1])].pt[1]
    
    srcImg_coord = [[x1,y1],[x2,y2],[x3,y3],[x4,y4]]
    replace_coord = [[x11,y11],[x22,y22],[x33,y33],[x44,y44]]

    return srcImg_coord, replace_coord

def myRANSAC(e,iteration,match,kp1,kp2):
    final_match = []
    k = 0
    while k < iteration:
        n1,n2,n3,n4 = np.random.randint(0,len(match),4) #choose 4 keypoint pairs
        left_coord, right_coord = choosepoint(kp1,kp2,match,n1,n2,n3,n4)
        H = getHomography_list(left_coord, right_coord)
        print('第',k,'次')  
        new_match = []        
        for i in range(len(match)):
            x1 = kp1[int(match[i][0])].pt[0]
            y1 = kp1[int(match[i][0])].pt[1]
            x2 = kp2[int(match[i][1])].pt[0]
            y2 = kp2[int(match[i][1])].pt[1]
            hx1 = (H[0][0]*x1 + H[0][1]*y1 + H[0][2]) / (H[2][0]*x1 + H[2][1]*y1 + H[2][2])
            hy1 = (H[1][0]*x1 + H[1][1]*y1 + H[1][2]) / (H[2][0]*x1 + H[2][1]*y1 + H[2][2])

            d = np.sqrt((x2 - hx1)**2 + (y2 - hy1)**2)
            if d < e:
                new_match.append(match[i])
        if len(new_match) > len(final_match):
            final_match = new_match
        k += 1

    #print('final_match',final_match)
    return final_match

def matchandcoord(img1, img2, match,kp1,kp2):
    img_src_coordinate = []
    replace_coordinate = []
    splicedImg = np.hstack((img1, img2))        #horizontally spliced img1 + img2
    for i in range(len(match)):
        x1 = kp1[int(match[i][0])].pt[0]
        y1 = kp1[int(match[i][0])].pt[1]
        x2 = kp2[int(match[i][1])].pt[0]
        y2 = kp2[int(match[i][1])].pt[1]
        img_src_coordinate.append([x1, y1])
        replace_coordinate.append([x2, y2])
        cv2.line(splicedImg,(int(x1),int(y1)),(int(x2)+img2.shape[1],int(y2)),(255,0,0),1)
    return splicedImg,img_src_coordinate,replace_coordinate

def getHomography_list(srcImg_coord, replace_coord):
    if len(srcImg_coord) != len(replace_coord):
        raise RuntimeError('number of points do not match, please choose 4 points')

    n = len(srcImg_coord)
    A = np.zeros((2*n,9))
    for i in range(0,n):
        A[2*i] = [srcImg_coord[i][0],srcImg_coord[i][1],1,0,0,0,
                  -replace_coord[i][0]*srcImg_coord[i][0],-replace_coord[i][0]*srcImg_coord[i][1],-replace_coord[i][0]]
        A[2*i+1] = [0,0,0,srcImg_coord[i][0],srcImg_coord[i][1],1,
                    -replace_coord[i][1]*srcImg_coord[i][0],-replace_coord[i][1]*srcImg_coord[i][1],-replace_coord[i][1]]
    U,S,V = linalg.svd(A) #SVD method
    H = V[8].reshape((3,3))
    
    return H / H[2,2]# uniformization

def getStitch(srcImg, dstImg, H, canvas_size, x0, y0, step):
    canvas = np.zeros(canvas_size)
    i=0
    while i < srcImg.shape[0]:
        j=0
        while j < srcImg.shape[1]:
            hx = (H[0][0]*j + H[0][1]*i + H[0][2]) / (H[2][0]*j + H[2][1]*i + H[2][2])
            hy = (H[1][0]*j + H[1][1]*i + H[1][2]) / (H[2][0]*j + H[2][1]*i + H[2][2])
            if int(hy)+x0 < canvas.shape[0] and int(hx)+y0 < canvas.shape[1]:
                canvas[int(hy)+x0][int(hx)+y0] = srcImg[int(i)][int(j)]
            j += step
        i += step
        
    for i in range(x0,x0+dstImg.shape[0]):
        for j in range(y0,y0+dstImg.shape[1]):
            canvas[i][j] = dstImg[i-x0][j-y0]    
    return canvas     

def getStitch2(srcImg, dstImg, H, canvas_size, x0, y0, step):
    canvas = np.zeros(canvas_size)
    i=0
    while i < srcImg.shape[0]:
        j=0
        while j < srcImg.shape[1]:
            x = j - 1280
            y = i - 280
            hx = (H[0][0]*x + H[0][1]*y + H[0][2]) / (H[2][0]*x + H[2][1]*y + H[2][2])
            hy = (H[1][0]*x + H[1][1]*y + H[1][2]) / (H[2][0]*x + H[2][1]*y + H[2][2])
            if 0<int(hy)+x0 < canvas.shape[0] and 0<int(hx)+y0 < canvas.shape[1]:
                canvas[int(hy)+x0][int(hx)+y0] = srcImg[int(i)][int(j)]
            j += step
        i += step
    for i in range(x0,x0+dstImg.shape[0]):
        for j in range(y0,y0+dstImg.shape[1]):
            canvas[i][j] = dstImg[i-x0][j-y0]
    return canvas   

def getStitch3(srcImg, dstImg, H, canvas_size, x0, y0, step):
    canvas = np.zeros(canvas_size)
    i=0
    while i < srcImg.shape[0]:
        j=0
        while j < srcImg.shape[1]:
            x = j - 2560
            y = i - 280
            hx = (H[0][0]*x + H[0][1]*y + H[0][2]) / (H[2][0]*x + H[2][1]*y + H[2][2])
            hy = (H[1][0]*x + H[1][1]*y + H[1][2]) / (H[2][0]*x + H[2][1]*y + H[2][2])
            if 0<int(hy)+x0 < canvas.shape[0] and 0<int(hx)+y0 < canvas.shape[1]:
                canvas[int(hy)+x0][int(hx)+y0] = srcImg[int(i)][int(j)]
            j += step
        i += step
    for i in range(x0,x0+dstImg.shape[0]):
        for j in range(y0,y0+dstImg.shape[1]):
            canvas[i][j] = dstImg[i-x0][j-y0]
    return canvas  
#=======================================
e = 0.5
iteration = 10000
ransac_kp_12 = myRANSAC(e, iteration, match12, kp1, kp2)
ransac_kp_23 = myRANSAC(e, iteration, match23, kp2, kp3)
ransac_kp_34 = myRANSAC(e, iteration, match34, kp3, kp4)

ransac_match_12,img_src_coord12,replace_coord12 = matchandcoord(img1, img2, ransac_kp_12,kp1,kp2)
ransac_match_23,img_src_coord23,replace_coord23 = matchandcoord(img2, img3, ransac_kp_23,kp2,kp3)
ransac_match_34,img_src_coord34,replace_coord34 = matchandcoord(img3, img4, ransac_kp_34,kp3,kp4)
H_best1 = getHomography_list(img_src_coord12, replace_coord12)
H_best2 = getHomography_list(img_src_coord23, replace_coord23)
H_best3 = getHomography_list(img_src_coord34, replace_coord34)

canvas1 = getStitch(img1, img2, H_best1, [1200,1920,3], 280, 1280, step = 0.1)
canvas2 = getStitch2(canvas1, img3, H_best2, [1600,3200,3], 280, 2560, step = 0.1)
canvas3 = getStitch3(canvas2, img4, H_best3, [2400,4800,3], 680, 4160, step = 0.1)
plt.figure(figsize=(20,20))
plt.subplot(311)
plt.imshow(canvas1.astype('uint8'))
plt.subplot(312)
plt.imshow(canvas2.astype('uint8'))
plt.subplot(313)
plt.imshow(canvas3.astype('uint8'))
plt.show()
```
