```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from gco import pygco 
import cv2

# def dist(c1, c2):
#     return (abs(c1[0]-c2[0])+abs(c1[1]-c2[1])+abs(c1[2]-c2[2]))/3

def dist(patch1, patch2):
    pixel_count = patch1.size
    _patch1 = patch1.flatten()
    _patch2 = patch2.flatten()
    value = 0.0
    for i in range(pixel_count):
        value += abs(float(_patch1[i]) - float(_patch2[i]))
    return value/pixel_count    

#=======================================================

leftImg = cv2.imread('Img/left2.png')
rightImg = cv2.imread('Img/right2.png')

leftImg = cv2.cvtColor(leftImg,cv2.COLOR_BGR2RGB)
rightImg = cv2.cvtColor(rightImg,cv2.COLOR_BGR2RGB)
newImg = leftImg

m_lambda = 0.05
num_label = 60

H = leftImg.shape[0] #height of img
W = leftImg.shape[1] #width of img

unary = np.zeros([H, W, num_label])
pairwise = (1 - np.eye(num_label)) * m_lambda

for y in range(0, H):
    for x in range(0, W):
        for i in range(0, num_label):
            c1 = leftImg[y][x]
            if (x-i) >= 0:
                c2 = rightImg[y][x-i]
            elif (x-i) < 0:
                c2 = rightImg[y][0]
            unary[y][x][i] = dist(c1, c2) / 255
 
n_labels = pygco.cut_grid_graph_simple(unary, pairwise, n_iter=-1)
n_labels = n_labels.reshape(H, W)

for y in range(0, H):
    for x in range(0, W):
        newImg[y][x] = n_labels[y][x] * 255 / num_label * np.array([1,1,1])
# for y in range(1, int(H/patchsize)):
#     for x in range(1, int(W/patchsize)):
#         newImg[(y-1)*patchsize : y*patchsize ,(x-1)*patchsize : x*patchsize, :] = int(n_labels[y-1][x-1]  * 255 / num_label) * np.ones([patchsize,patchsize,3])
        
plt.imshow(newImg)
plt.show()
```
