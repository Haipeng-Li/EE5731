```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from gco import pygco 
import cv2
import os

# load the cameras matrix in cameras.txt
framenum = np.loadtxt('Road/Road/cameras.txt',dtype=np.float32, max_rows=1)
cam = np.loadtxt('Road/Road/cameras.txt',dtype=np.float32, skiprows=1)
framenum = int(framenum)
K = []
R = []
T = []
for i in range(0, framenum):
    K.append(cam[7*i:7*i+3, :])
    R.append(cam[7*i+3:7*i+6, :])
    T.append(cam[7*i+6, :])

K = np.array(K)
R = np.array(R)
T = np.array(T)

# load images
path = 'Road/Road/src'
datanames = os.listdir(path)
img = [] 
img_v = []
for i in datanames:
    tmp = cv2.imread('Road/Road/src/'+str(i))
    tmp = cv2.cvtColor(tmp,cv2.COLOR_BGR2RGB)
    tmp_v = tmp.reshape(tmp.shape[0]*tmp.shape[1], tmp.shape[2])
    img.append(tmp)
    img_v.append(tmp_v)
img = np.array(img) #(141, 540, 960, 3)
img_v = np.array(img_v)
```

```python
# parameters
src = 140
frames = 5 # the number of projected frame
sigma_c = 10
sigma_d = 1 # standard deviation标准差
D = np.linspace(0.0001, 0.01, 50) # disparity
H, W, channel = img[0].shape
```

# Disparity initialization

```python
#  photo consistency constraint
unary = np.zeros([H, W, len(D)])

for i in range(135, 135+frames):
    print('frame ',i)
    K1 = K[src]
    R1 = R[src]
    T1 = T[src]
    K2 = K[i]
    R2 = R[i]
    T2 = T[i]
    term1 = K2 @ R2 @ R1.T @ np.linalg.inv(K1)
    term2 = K2 @ R2 @ np.mat(T1-T2).T
    img1 = img[src]
    img2 = img[i]
    unary_c = np.zeros([H, W, len(D)])
    for y in range(H):
        for x in range(W):
            coord1 = np.array([[x],[y],[1]])
            for d in range(0, len(D)):

                coord2 = np.array(term1 @ coord1 + D[d] * term2)
                coord2 = np.round(coord2/coord2[2])
                
                if 0<=coord2[1]<H and 0<=coord2[0]<W:
                    consis = sum(abs(img1[y, x, :]-img2[int(coord2[1]),int(coord2[0]),:]))/255
                    #pc += (sigma_c / (sigma_c + consis))    
                    unary_c[y, x, d] = consis
    unary += unary_c
print('Initialization finished!')                           
```

```python
# normalization data term
dataterm = sigma_c / (sigma_c + unary)
# norm = np.zeros([H, W])
# for y in range(H):
#     for x in range(W):
#         norm[y][x] = np.max(unary[y][x])
#         dataterm[y][x] = 1 - dataterm[y][x] / norm[y][x]
norm = np.max(unary)
dataterm = 1 - dataterm / norm

# prior term
ita = 0.05*(D[-1] - D[0])
fp = np.zeros([50, 50])
for y in range(50):
    for x in range(50):
        fp[y][x] = np.sqrt((D[x]-D[y])**2)
        if fp[y][x]>ita:
            fp[y][x] = ita

```

```python
# weighting factor lambda
epsilon = 50
ws = 20/(D[-1] - D[0])
img1 = img[0]
lambda_v = np.zeros([H-1, W])
lambda_h = np.zeros([H, W-1])
Ux = np.zeros([H, W])
for y in range(H):
    for x in range(W):
        point = []
        if y-1>=0:
            point.append(img1[y-1][x]) 
        if y+1<H:
            point.append(img1[y+1][x]) 
        if x+1<W:
            point.append(img1[y][x+1])
        if x-1>=0:
            point.append(img1[y][x-1])
            
        point = np.array(point)
        Nx = point.shape[0]
        Unorm = 0
        for j in range(Nx):
            Unorm += 1 / (sum(abs(img1[y][x] - point[j])) + epsilon)
        Ux[y][x] = Nx/Unorm
for y in range(H-1):
    for x in range(W):          
        point_v = img1[y-1][x]
        lambda_v[y][x] = Ux[y][x] * ws / (sum(abs(img1[y][x] - point_v)) + epsilon)
for y in range(H):
    for x in range(W-1):     
        point_h = img1[y][x-1]
        lambda_h[y][x] = Ux[y][x] * ws / (sum(abs(img1[y][x] - point_h)) + epsilon)

lambda_v = lambda_v/np.max(lambda_v)
lambda_h = lambda_h/np.max(lambda_h)
```

```python
# n_labels = pygco.cut_grid_graph_simple(dataterm, pairwise, n_iter=-1)
n_labels = pygco.cut_grid_graph(dataterm, fp*100, lambda_v, lambda_h)
n_labels = n_labels.reshape(H, W)        
```

```python
def get_terms(obj_id, frame_id):
    K1 = K[obj_id]
    R1 = R[obj_id]
    T1 = T[obj_id]
    K2 = K[frame_id]
    R2 = R[frame_id]
    T2 = T[frame_id]
    term1 = K2 @ R2 @ R1.T @ np.linalg.inv(K1)
    term2 = K2 @ R2 @ np.mat(T1-T2).T

    return term1, term2
```
