```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from gco import pygco 
import cv2

def dist(c1, c2):
    return (abs(c1[0]-c2[0])+abs(c1[1]-c2[1])+abs(c1[2]-c2[2]))/3

srcImg = cv2.imread('Img/noisyimg.jpg')
denoiseImg = srcImg
srcImg = cv2.cvtColor(srcImg,cv2.COLOR_BGR2RGB)
denoiseImg = cv2.cvtColor(denoiseImg,cv2.COLOR_BGR2RGB)

H = srcImg.shape[0] #height of img
W = srcImg.shape[1] #width of img


src_color = np.array([0,0,255]) # blue = foreground
sink_color = np.array([245,210,110]) #yellow = background

unary = np.zeros([H, W, 2])
pairwise = (1 - np.eye(2))

for y in range(0, H):
    for x in range(0, W):
        c = srcImg[y][x]
        src_label = dist(c, src_color)/255
        sink_label = dist(c, sink_color)/255
        unary[y][x] = [src_label, sink_label]
        
m_lambda = 1
labels = pygco.cut_grid_graph_simple(unary, pairwise * m_lambda, n_iter=-1)
labels = labels.reshape(H, W)

for y in range(0, H):
    for x in range(0, W):
        if labels[y][x] == 1:
            denoiseImg[y][x] = sink_color
        if labels[y][x] == 0:
            denoiseImg[y][x] = src_color
plt.figure(figsize=(10,10))           
plt.subplot(211)
plt.title('Original image')
plt.imshow(srcImg)
plt.subplot(212)
plt.title('m_lambda = 1')
plt.imshow(denoiseImg)
plt.show()
```

```python
m_lambda = 0.1
labels = pygco.cut_grid_graph_simple(unary, pairwise * m_lambda, n_iter=-1)
labels = labels.reshape(H, W)

for y in range(0, H):
    for x in range(0, W):
        if labels[y][x] == 1:
            denoiseImg[y][x] = sink_color
        if labels[y][x] == 0:
            denoiseImg[y][x] = src_color
plt.figure(figsize=(5,5))           
plt.title('m_lambda = 0.1')
plt.imshow(denoiseImg)
plt.show()
```

```python
m_lambda = 3
labels = pygco.cut_grid_graph_simple(unary, pairwise * m_lambda, n_iter=-1)
labels = labels.reshape(H, W)

for y in range(0, H):
    for x in range(0, W):
        if labels[y][x] == 1:
            denoiseImg[y][x] = sink_color
        if labels[y][x] == 0:
            denoiseImg[y][x] = src_color
plt.figure(figsize=(5,5))           
plt.title('m_lambda = 3')
plt.imshow(denoiseImg)
plt.show()
```

```python
m_lambda = 5
labels = pygco.cut_grid_graph_simple(unary, pairwise * m_lambda, n_iter=-1)
labels = labels.reshape(H, W)

for y in range(0, H):
    for x in range(0, W):
        if labels[y][x] == 1:
            denoiseImg[y][x] = sink_color
        if labels[y][x] == 0:
            denoiseImg[y][x] = src_color
plt.figure(figsize=(5,5))           
plt.title('m_lambda = 5')
plt.imshow(denoiseImg)
plt.show()
```

### After the test of different lambda values, the best on is when lambda = 1.
### We can find that when lambda is small, like the value 0.1, some noise can not be removed. When lambda increasing, the image can be denoised well, and the edge can also be softened. While when lambda is too large, like the value 5, some information will be mistakenly removed.
